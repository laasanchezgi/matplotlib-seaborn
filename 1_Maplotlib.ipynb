## Subplots

````python
x = np.linspace(0,5,11)
y = np.sin(x)
````

````python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'b')
````

````python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'r')
ax2.plot(y,x,'b')
````

````python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), 'r')
axes[0,2].plot(x, np.tan(x), 'y')
axes[0,3].plot(x, np.sinh(x), 'pink')
axes[1,0].plot(x, np.cosh(x))
axes[1,1].plot(x, np.tanh(x))
axes[1,2].plot(x, np.cos(x)**2)
axes[1,3].plot(x, np.cosh(x)**2)
fig.tight_layout()
````

## Leyendas, etiquetas, t√≠tulos, tama√±o

````python
x = np.linspace(0,5,11)
y = np.sin(x)
````

````python
fig, axes = plt.subplots(1,2, figsize=(10,5))

axes[0].set_title('Relacion X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].plot(x,y, label='$sin(x)$')
axes[0].legend()

axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].plot(y,x, label='$sin(y)$')
axes[1].legend()
````

````python
plt.plot(x,y, label = 'sin(x)')
plt.title('Grafica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.5))
````

## Bar plot

````python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000,800,900,100,300]
````

````python
plt.bar(country, population)
plt.show()
````

````python
plt.bar(country, population, width=0.5, color=['red','blue'])
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'colombia', 'Alemania'), rotation=45)
plt.show()
````

````python
plt.barh(country, population)
plt.show()
````

## Crea otro tipo de gr√°ficas

````python
data = np.random.randint(1,50,100)
data
````

````python
plt.hist(data, bins=15, width=2.5, histtype='bar')
plt.show()
````

````python
plt.hist(data, bins=15, histtype='step')
plt.show()
````

````python
plt.boxplot(data)
plt.show()
````

````python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
````

````python
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
````

````python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
````
